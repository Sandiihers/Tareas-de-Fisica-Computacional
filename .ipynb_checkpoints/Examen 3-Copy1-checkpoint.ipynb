<span style="color:blue">1. Ecuentra la soluci√≥n n√∫merica para la ecuaci√≥n diferencial $\nabla^{2}\phi=0$ en dos dimensiones con condiciones a la frontera $\phi(0,y)=\phi(x,0)=\phi(1,y)=\phi(x,1)0$ y  $\phi=1$ para todos los puntos x,y en dos c√≠rculos de radio $1/4$ con centros en (0.4,0.5) y (0.6,0.5) (3 puntos)<span>
    

Tal como se hizo en el Notebook 14 comenzamos programando la forma de generar la malla num√©rica tomando en cuenta los condiciones de las fronteras.

```julia
function Malla_eje1(x,y,r,V_0=1.0)
    Mat=zeros(length(x),length(y))
    for i in 2:length(y)-1 #columnas
        for j in 2:length(x)-1 #renglones
            #Condiciones a la frontera
            #dos c√≠rculos
            if sqrt((x[j]-0.4)^2+(y[i]-0.5)^2 )<=r || sqrt((x[j]-0.6)^2+(y[i]-0.5)^2 )<=r
                Mat[i,j]=V_0
            else
                Mat[i,j] = rand()
            end
        end
    end
    return Mat
end
```

```julia
x=collect(0.0:0.01:1.0)
y=collect(0.0:0.01:1.0)
mi_malla=Malla_eje1(x,y,0.25) #1/4 de radio
```

Gr√°ficamos nuesra malla

```julia
heatmap(x,y,mi_malla)
```

La malla cumple con las condiciones de frontera

Ahora usamos el m√©todo de Jacobi definido como:
$$\phi_{i,j}=\frac{1}{4}\left(\phi_{i+1,j}+\phi_{i-1,j}+\phi_{i,j+1}-2\phi_{i,j}+\phi_{i,j-1}\right)$$

```julia
function relax!(Mat,V_0=1.)
    for i in 2:length(y)-1 # columnas 
        for j in 2:length(x)-1 # renglones
            if Mat[i,j] ‚â† V_0
                Mat[i,j] = 0.25*( Mat[i+1,j] + Mat[i-1,j] + Mat[i,j+1] + Mat[i,j-1] )
            end
        end
    end
    return Mat
end
```

```julia
mi_malla2=relax!(mi_malla)
```

Ahora implementamos el m√©todo anterior hasta alcanzar convergencia en el m√©todo num√©rico. Para ello definimos una nueva funci√≥n; repitiendo la funci√≥n ***relax!*** hasta que se alcance la convergencia.

```julia
function metodo_relax_v1!(Mat,error=1e-5)
    testigo,cuenta = true,0
    while testigo == true
        Mat_old = copy(Mat)
        Mat = relax!(Mat)
        cuenta += 1
        if maximum( abs.( Mat .- Mat_old) ) <= error
            testigo=false
        end
    end
    return Mat,cuenta
end
```

Ahora usamos nuestra funci√≥n **metodo_relax_v1!** en **Malla_eje1**

```julia
x_2=collect(0:0.01:1)
y_2=collect(0:0.01:1)
#Definimos a phi como malla usando "Malla_eje1"  y "x_2" y "y_2"
phi=Malla_eje1(x_2,y_2,0.25) #1/4 de radio
phi,num=metodo_relax_v1!(phi)
phi
```

<span style="color:blue">Dibuje las l√≠neas de campo vectorial que se generan de obtener $-\nabla\phi$. (1 punto extra)<span>

Para sacar el campo el√©ctrico se define una funci√≥n que regrese la gradiente

Ser√°n dos funciones una para x y otra para y, ya que esta restringido a (xy)

```julia
function gradienteenx(x,y,phi)
     h=x[2]-x[1]
    gradiente=zeros(length(x),length(y))
    for i in 2:length(y)-1
        for j in 2:length(x)-1
            gradiente[i,j]=(phi[i,j+1]-phi[i,j-1])/h
        end
    end
    return gradiente
end
    
```

```julia
function gradienteeny(x,y,phi)
    h=y[2]-y[1]
    gradiente=zeros(length(x),length(x))
    for i in 2:length(y)-1
        for j in 2:length(x)-1
            gradiente[i,j]=(phi[i+1,j]-phi[i-1,j])/h
        end
    end
    return gradiente
end
```

```julia
Ex=-1 .*gradienteenx(x_2,y_2,phi) #‚àí‚àáùúô
Ey=-1 .*gradienteeny(x_2,y_2,phi) #‚àí‚àáùúô
p3=streamplot(x_2,y_2,Ex,Ey,density=2,color="k")
#plot!(p3,title="h",xlabel="x",ylabel="y")
"L√≠neas de campo vectorial que se generan de obtener ‚àí‚àáùúô"
```

<span style="color:blue">Puede suponer que el coeficiente de difusi√≥n es 1<span>

En el  Nootebook 15 M√©todo de Crank-Nicholson se aplica a Psi tomando solo los puntos interiores de esta. (Psi es un malla). 
entonces empezamos a definir una malla que cumpla con las condiciones de frontera.

```julia
function Psi_ini_2circ(x,y) 
    Nx,Ny = length(x),length(y)
    Œ® = zeros(Ny,Nx)
    for i in 2:Nx-1
        for j in 2:Ny-1
            if sqrt( (x[i]-0.4)^2 + (y[j]-0.5)^2 ) <= 0.25 || sqrt( (x[i]-0.6)^2 + (y[j]-0.5)^2 ) <= 0.25
                Œ®[j,i] = 1.0
            end
        end
    end
    return Œ®
end
```

Definimos una funci√≥n que implementa al M√©todo de Crank-Nicholson (para obtener r)

```julia
function Mat_evol_CN_2D(x,y,t,D=1.0)
    Nx,Ny = length(x),length(y)
    Œîx,Œîy,Œît = x[2]-x[1],y[2]-y[1],t[2]-t[1]
    r = (D*Œît)/(2*Œîx*Œîy) 
    println("r = ",r)
    A = zeros((Nx-2)*(Ny-2),(Nx-2)*(Ny-2))
    B = zeros((Nx-2)*(Ny-2),(Nx-2)*(Ny-2))
    # Elementos de la diagonal 
    for i in 1:Nx-2
        for j in 1:Ny-2
            A[(Ny-2)*(i-1)+j,(Ny-2)*(i-1)+j] = 1+4*r
            B[(Ny-2)*(i-1)+j,(Ny-2)*(i-1)+j] = 1-4*r
        end
    end
    # Elementos de la diagonal superior e inferior de la diagonal principal
    for i in 1:Nx-2
        for j in 1:Ny-3
            A[(Ny-2)*(i-1)+j,(Ny-2)*(i-1)+j+1] = -r
            A[(Ny-2)*(i-1)+j+1,(Ny-2)*(i-1)+j] = -r
            B[(Ny-2)*(i-1)+j,(Ny-2)*(i-1)+j+1] = r
            B[(Ny-2)*(i-1)+j+1,(Ny-2)*(i-1)+j] = r
        end
    end
    # Elementos diagonales de la diagonal matricial superior e inferior
    for i in 1:Nx-3
        for j in 1:Ny-2
            A[(Ny-2)*(i-1)+j,(Ny-2)*(i)+j] = -r
            A[(Ny-2)*(i)+j,(Ny-2)*(i-1)+j] = -r
            B[(Ny-2)*(i-1)+j,(Ny-2)*(i)+j] = r
            B[(Ny-2)*(i)+j,(Ny-2)*(i-1)+j] = r
        end
    end
    A_inversa = inv(A)
    Evol = A_inversa*B
    return Evol
end
```

M√©todo de Crank-Nicholson

```julia
function Crank_Nicholson_2D(Œ®‚ÇÄ,M_evo,t)
    Nx,Ny = length(Œ®‚ÇÄ[:,1]),length(Œ®‚ÇÄ[1,:])
    Œ®_xy_t = zeros(length(Œ®‚ÇÄ[:,1]),length(Œ®‚ÇÄ[1,:]),length(t))
    Œ®_xy_t[2:end-1,2:end-1,1] = Œ®‚ÇÄ[2:end-1,2:end-1]
    Œ®_in_t = reshape( Œ®‚ÇÄ[2:end-1,2:end-1] , ((Nx-2)*(Ny-2)),1 )
    for i in 2:length(t)
        Œ®_in_t = M_evo*Œ®_in_t
        Œ®_xy_t[2:end-1,2:end-1,i] = reshape( Œ®_in_t , (Nx-2,Ny-2) )
    end
    return Œ®_xy_t
end
```

Ahora aplicamos nuestra funci√≥n implementaria del M√©todo de Crank-Nicholson para obtener r

<span style="color:blue">3. Encuentre los estados estacionarios de la ecuaci√≥n de onda en 2D para una membrana rectangular $L_{x}=\pi$, $L_{y}=2\pi$. Compare la soluci√≥n num√©rica de los eigenvalores con la soluci√≥n anal√≠tica. (3 puntos)<span>

<span style="color:blue">¬øQu√© sucedde con la soluci√≥n si ahora $L_{x}=\pi$, $L_{y}=\sqrt{2\pi}$?<span>

Definimos una funci√≥n para encontrar los eigenvalores de la ecuaci√≥n de onda

```julia
function Mat_onda_2D(x,y)
    Nx,Ny = length(x),length(y)
    A = zeros(Ny*Nx,Ny*Nx)
    for i in 1:Nx
        for j in 1:Ny
            A[Ny*(i-1)+j,Ny*(i-1)+j] = -4
        end
    end
    for i in 1:Nx
        for j in 1:Ny-1
            A[Ny*(i-1)+j,Ny*(i-1)+j+1] = 1
            A[Ny*(i-1)+j+1,Ny*(i-1)+j] = 1
        end
    end
    for i in 1:Nx-1
        for j in 1:Ny
            A[Ny*(i-1)+j,Ny*i+j] = 1
            A[Ny*i+j,Ny*(i-1)+j] = 1
        end
    end
    return A
end
```

Obtenemos los eigenvalores de manera num√©rica correspodientes a la membrana de 2$\pi$ por $\pi$.

```julia
#Definimos los puntos x, y de la menbrana
x=collect(0:0.04:pi)
y=collect(0:0.04:(2*pi))
h=x[2]-x[1] # hx=hy=h
#Definimos una Matriz que ser√° la matriz de la cual obtendremos los eigenvalores (usando "x" y "y")
M=Mat_onda_2D(x[2:end-1],y[2:end-1])
#Obtenemos los eigenvalores de la Matriz definida anteriormente
vals = eigvals(M)
#Calculamos los nodos de vibraci√≥n pedidos 
valores=sort(sqrt.(-vals)/(h)) #usamos sort para ordenar 
```

Ahora debemos obtener los eigenvalores de manera anal√≠tica.

Los eigenvalores de una membrana (manera anal√≠tica) est√°n dados por la ecuaci√≥n:
$$\lambda_{nm}=c\pi\sqrt{{\frac{m^2}{a^2}}+{\frac{n^2}{b^2}}} $$

con $m,n \in  \mathbb{Z}$ y $a$=ancho de la membrana, $b$=largo de la membrana.

```julia
#Definimos las n's y m's que debem ser enteros
n=collect(1:1:3)
m=collect(1:1:3)
#El ancho y largo de la membrana
a=œÄ
b=2œÄ
#La funci√≥n que nos de el eigenvalor correspondiente a n y m
f(n,m)=œÄ*sqrt.(m^2/a^2+n^2/b^2)
#Definimos un arreglo vac√≠o al cual iremos anexando los eigenvalores
Œª=[]
#Realizamos un ciclo para obtener los eigenvalores para cada n y m
for i in n
    for j in m
       push!(Œª,f(i,j))
    end
end
#Regresamos los eigenvalores obtenidos y los ordenamos de menor a  mayor
return sort(Œª)
```

Obtenemos el valor absoluto de la diferencia entre los eigenvalores obtenido de manera anal√≠tica y los obtenidosde manera n√∫merica

```julia
difabs1=broadcast(abs, sort(Œª)-valores[1:length(Œª)])
```

```julia
dif=sort(Œª)-valores[1:length(Œª)]
```

```julia
difabs=broadcast(abs, dif)
```

Hacemos una tabla para comparar los resultados 

```julia
data=DataFrame(Anal√≠tico=sort(Œª),Num√©rico=valores[1:length(Œª)],Diferencia=difabs)
```
